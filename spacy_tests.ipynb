```python
import spacy
```

```python
nlp = spacy.load("fr_core_news_sm")
```

```python
text = "je veux aller de Paris à Marseille"
```

```python
doc = nlp(text)
```

```python
for token in doc:
    if token.dep_ == "prep":
        print(f"Préposition : {token.text} → {token.head.text}")
```

```python
for token in doc:
    print(f"{token} -> {token.dep_}")
```
